# Chi-Square Test

Null Hypothesis: The features categories have no effect on Target Variable

Alternate Hypothesis: one or more features have significant effect on target variable

```python
from sklearn.feature_selection import chi2
```

```python
x_ch=cc_ml_knn.drop(['label'],axis=1)
```

```python
y_ch=cc_ml_knn['label']
```

```python
cc_ml.columns
```

```python
chi_scores = chi2(x_ch,y_ch)
```

```python
chi_scores
```

here first array represents chi square values and second array represnts p-values

```python
p_values = pd.Series(chi_scores[1],index = x_ch.columns)
p_values.sort_values(ascending = False , inplace = True)
```

```python
p_values.plot.bar()
```

High P-value for columns indicate these columns are not dependent on Target Variable

We select best features required for evaluation of the model

# Model Building

## Getting Independent features in dataset x and dependent features in y

```python
x=cc_ml_knn.drop(['label'],axis=1)
x
```

```python
y=cc_ml_knn['label']
y
```

## Splitting data into Training and Test data

```python
### Machine Learning libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, fbeta_score
```

```python
### random state train test split will be same with all people using random_state=1
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=1)
```

# Models

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier # Importing Decision Tree Classifier
```

```python
# Create Decision Tree classifer object
classification = DecisionTreeClassifier(max_depth=3)

# Train Decision Tree Classifer
classification = classification.fit(x_train,y_train)

#Predict the response for test dataset
y_pred_dc= classification.predict(x_test)
```

```python
# Model Accuracy, how often is the classifier correct?
# print classification report
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred_dc))
```

```python
cf_mat_dc=confusion_matrix(y_test, y_pred_dc)
cf_mat_dc
```

```python
### accuracy using accuracy_score
print(round(accuracy_score(y_test,y_pred_dc), 2))
```

```python
print(precision_score(y_test, y_pred_dc))
```

```python
print(round(recall_score(y_test, y_pred_dc), 2))
```

```python
from sklearn.metrics import f1_score
print(round(f1_score(y_test, y_pred_dc), 2))
```
